```python
'''
Exercise 1: Prime Numbers
Write a Python program that checks whether a given number is prime or not.
 A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.
'''
```

```python
# Taking input from the user and storing it in variable x.
x = int(input("Enter a Number:"))

# If the number is lessthan or equalto one then it prints not a prime.
if (x <= 1): 
    print("It is not a Prime Number.")
else:
    # Assuming the give number is prime.
    is_prime = True

# cheking the number is a prime number or not. 
for i in range(2, x):
    if (x % i == 0):
        is_prime = False
        break # break the condition.

if is_prime:
    print("It is a Prime Number") # It prints if it is true.
else:
    print("It is not a Prime Number.") # It prints if it is not true.
```

```python
'''
Exercise 2: Product of Random Numbers
  Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers.
The program should then check if the user's answer is correct and display an appropriate message.
'''
```

```python
import random
```

```python
# a and b are random number.
a = random.randint(2, 50)
b = random.randint(49, 100)

# It prints two random numbers.
print(a, "*", b)

# Product of two random numbers.
c = a * b

# user will give answer.
answer = int(input("Enter the answer:"))

# Assuming answer is false
is_answer = False

# If answer matches is_answer changes to true.
if (answer == c):
    is_answer = True

if is_answer:
    print("Correct Answer. Well done!") # it prints if it is true.
else:
    print("Wrong Answer! The correct answer is", c) # It prints if it is not true.
```

```python
'''
Exercise 4:
✅ Problem Statement (Question): Hospital Billing System
A hospital charges its patients based on the following:
•	Room charges per day: ₹2,000
•	Doctor consultation fee (flat): ₹1,500
•	Lab test charges: Based on the number of tests taken. Each test costs ₹300
•	Medicine charges: Total cost of medicines provided
•	Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied
Write a Python program to calculate the final bill for a patient who:
•	Stayed for 4 days
•	Had 3 lab tests
•	Had medicine charges of ₹2,400
'''
```

```python
# input variables.
room_Charges_Per_Day = 2000
doctor_Consultation_Fee = 1500
lab_Test_Charges = 300
discount = 10

# user inputs
no_Of_Days_Stayed = int(input("Enter No-of Days Stayed:"))
no_Of_Lab_Tests = int(input("Enter No-of Lab Tests:"))
medicine_Charges = int(input("Total Medicine Charges:"))

# Calculate the Total bill
room_Charges = room_Charges_Per_Day * no_Of_Days_Stayed
lab_Test_Charges = lab_Test_Charges * no_Of_Lab_Tests
total_Bill = doctor_Consultation_Fee + lab_Test_Charges + room_Charges + medicine_Charges

if (total_Bill <= 10000):
    print()
    print("Room Charges (4 days)", room_Charges)
    print("Doctor Consultation Fee:", doctor_Consultation_Fee)
    print("Lab Test Charges (3 tests):", lab_Test_Charges)
    print("Medicine Charges:", medicine_Charges)
    print()
    print("Total Bill:", total_Bill, "RS") # It prints if total bill is lessthan 10000.
else:
    discount_Amount = (total_Bill * discount) / 100
    total_Bill = total_Bill - discount_Amount

    print()
    print("Room Charges (4 days)", room_Charges)
    print("Doctor Consultation Fee:", doctor_Consultation_Fee)
    print("Lab Test Charges (3 tests):", lab_Test_Charges)
    print("Medicine Charges:", medicine_Charges)
    print("Discount Amount:", discount_Amount)
    print()
    print("Total Bill:", total_Bill, "RS") # It prints if total bill is exceeds 10000.
```
