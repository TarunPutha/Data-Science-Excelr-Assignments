 1. Create a list of 5 integers. 
    Perform and print the result of the following operations: 
    append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# Creating a list of 5 integers.
nums1 = [10, 20, 30, 40, 50]
print(nums1)

# append(value)
nums1.append(60) # It adds an element to the last index of the list.
print(nums1)

# extend([values])
nums1.extend([70, 80]) # It add an multiple elements to the end of the list.
print(nums1)

# insert(index, value)
nums1.insert(0, 5) # It adds an element to the 0th index.
print(nums1)

# remove(value)
nums1.remove(50) # It removes value from the list.
print(nums1)

# pop
nums1.pop() # It removes last index from the list.
print(nums1)

# index(value)
index_vlaue = nums1.index(60) # It gives value index from the list
print(index_vlaue)

# count
cnt = nums1.count(60) # It counts how many noof times the value occurs in the list.
print(cnt)

# reverse
nums1.reverse() # It will reverse the list.(Descending Order)
print(nums1)

# sort
nums1.sort() # It will sort the list.(Ascending Order)
print(nums1)

# clear
nums1.clear() # It removes all elements from the list.
print(nums1)
```

7.	Given a number, check if it's positive, and if it is also even.
        If not positive, print if it's zero or negative.

```python
# Asking user to enter a number.
num = int(input("Enter a number:"))

# Checking positive or even positive or negative or zero.
if (num > 0):
    if (num % 2 == 0):
        print("It is Even Positive Number.")
    else:
        print("It is Positive Number.")
elif (num < 0):
    print("It is Negative Number.")
else:
    print("It is Zero.")
```

11. Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
        - Use .info(), .describe()
        - Select 'Name' and 'City' columns
        - Drop 'City' column
        - Fill any missing values in 'Salary' column with the mean
        - Remove any duplicate rows

```python
import numpy as np
import pandas as pd
```

```python
# Creating a DataFrame.
data = {
    'Name' : ['Tarun', 'Sonu', 'Mohith', 'Tarun', 'Krishna', 'Radha', 'Rama', 'Seetha', 'Shiva', 'Parvathi'],
    'Age' : [21, 22, 21, 21, 35, 32, 54, 53, 73, 70],
    'City' : ['Rajampet', 'Bangarupalyam', 'Nellore', 'Rajampet', 'Dwaraka', 'Dwaraka', 'Tirupati', 'Arunachalam', 'Kaasi', 'Kailasam'],
    'Salary' : [15000, np.nan, 20000, 15000, np.nan, 50000, 60000, 70000, 80000, 90000]
}

df1 = pd.DataFrame(data)
df1
```

```python
df1.info()
```

```python
df1.describe()
```

```python
# Selecting columns.
df1[['Name', 'City']]
```

```python
# Droping column.
df2 = df1.drop(columns = ['City'])
df2
```

```python
# Finding any missing values
print('Missing Values in Salary column:', df2["Salary"].isnull().sum())

# Finding mean for salary column.
salary_Mean = df2['Salary'].mean()

# Filling mean for missing values in salary column.
df2['Salary'].fillna(salary_Mean, inplace = True)
df2
```

```python
# Removing duplicate rows in dataframe.
df2.drop_duplicates(inplace = True)
df2
```
